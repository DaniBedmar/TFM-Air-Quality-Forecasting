## Notebook for visualizing Introduction data

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
import os
```

### The source of air pollution

```python
labels     = ['Road','Trains','Waterborne','Airborne']
pollutants = ['NOx','PM2.5','PM10','CO']

nox  = np.array([47.2,1.1,38.2,13.5])
pm25 = np.array([50.93,3.616,3.27,42.178])
pm10 = np.array([60.49,2.628,4.62,32.26])
co   = np.array([84.578,8.733,0.35,6.338])

data = np.vstack([nox, pm25, pm10, co])
colors = ['mediumseagreen','gold', 'skyblue', 'plum' ]  
x = np.arange(len(pollutants))
width = 0.6

fig, ax = plt.subplots()
bottom = np.zeros(len(pollutants))

for i, mode in enumerate(labels):
    ax.bar(x,
           data[:, i],
           width,
           bottom=bottom,
           label=mode,
           color=colors[i],
           zorder=2)
    bottom += data[:, i]

for spine in ax.spines.values():
    spine.set_visible(False)
ax.axhline(0, color='black', linewidth=0.5, zorder=1, alpha=0.5)
ax.yaxis.set_major_locator(plticker.MultipleLocator(10))
ax.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, zorder=1)

ax.set_xticks(x)
ax.set_xticklabels(pollutants)
ax.set_xlabel("Pollutant")
ax.set_ylabel("Percentage of Total Emissions")
ax.set_title("Transport Emission Breakdown by Pollutant")
ax.set_ylim(0, 105)
ax.legend(title="Transport Mode", bbox_to_anchor=(1.02, 0.7), borderaxespad=0)

plt.tight_layout()
image_path = os.path.join('..','Figures','Transport_emitors_by_pollutant.png')
plt.savefig(image_path)
plt.show()
```
